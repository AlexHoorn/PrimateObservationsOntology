```python
import pandas as pd
```

```python
inat_obs_df = pd.read_csv('observations_mapped.csv',sep=',')
inat_obs_df.head()
```

```python
wd_df = pd.read_csv('ids.csv',sep=',')
wd_df = wd_df.fillna(0)
wd_df = wd_df.astype({'iNaturalist_taxon_ID':'int64','GBIF_taxon_ID':'int64','ITIS_TSN':'int64','NCBI_taxonomy_ID':'int64'})
wd_df.head()
```

```python
NCBI_taxon_ID = []
iNat_taxon_ID = []
GBIF_taxon_ID = []
for item in inat_obs_df['taxon_id']:
    iNat_taxon_ID.append(item)
    row = wd_df.loc[wd_df['iNaturalist_taxon_ID'] == item]
    if(not row.empty):
        NCBI_taxon_ID.append(row['NCBI_taxonomy_ID'].values[0])
        GBIF_taxon_ID.append(row['GBIF_taxon_ID'].values[0])
    else:
        NCBI_taxon_ID.append(int(0))
        GBIF_taxon_ID.append(int(0))

print(len(iNat_taxon_ID))
print(len(NCBI_taxon_ID))
print(len(GBIF_taxon_ID))

    
```

```python
inat_obs_df["iNat_Taxon_ID"] = iNat_taxon_ID
inat_obs_df["NCBI_Taxon_ID"] = NCBI_taxon_ID
inat_obs_df["GBIF_Taxon_ID"] = GBIF_taxon_ID
```

```python
inat_obs_df.drop("taxon_id",axis=1,inplace=True)
```

```python
inat_obs_df.to_csv("obs_data_taxon_ID.csv",sep=',',index=False)
```
